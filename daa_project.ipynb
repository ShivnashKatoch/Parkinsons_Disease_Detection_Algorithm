Attribute information:

Matrix column entries (attributes):

name - ASCII subject name and recording number

MDVP:Fo(Hz) - Average vocal fundamental frequency

MDVP:Fhi(Hz) - Maximum vocal fundamental frequency

MDVP:Flo(Hz) - Minimum vocal fundamental frequency

MDVP:Jitter(%), MDVP:Jitter(Abs), MDVP:RAP, MDVP:PPQ, Jitter:DDP - Several measures of variation in fundamental frequency

MDVP:Shimmer,MDVP:Shimmer(dB),Shimmer:APQ3,Shimmer:APQ5,MDVP:APQ,Shimmer:DDA - Several measures of variation in amplitude

NHR, HNR - Two measures of the ratio of noise to tonal components in the voice
status - The health status of the subject (one) - Parkinson's, (zero) - healthy
RPDE, D2 - Two nonlinear dynamical complexity measures

DFA - Signal fractal scaling exponent

spread1,spread2,PPE - Three nonlinear measures of fundamental frequency variation

Data Pre-Processing

Separating the features(other column than status) & Target(Nothing but status column)

```python
X = parkinsons_data.drop(columns=['name','status'], axis=1) #for droping a column you use axis = 1 and for droping a row axis = 0 
Y = parkinsons_data['status']
```

```python
print(X)#it does not contain status of patiten
```

```python
print(Y)
```

Splitting the data to training data & Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
print(X.shape, X_train.shape, X_test.shape)

#here x data will be split in train and test data x_train represent training data and is corresponding label is store in y_train same for x_test
#here we use 20% data as testing data and 80% as training data  
```

Data Standardization

```python
# we use data standardization method where all the data will be transformed to a same range but it won't change the meaning of the data
scaler = StandardScaler()
```

```python
scaler.fit(X_train)#here fit function is used to fit this data in standard scaler
#we are not fitting x_test data here because we don't want our model to know the test data before hand
```

```python
X_train = scaler.transform(X_train)

X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

Model Training 

Support Vector Machine Model

```python
model = svm.SVC(kernel='linear')
#.SVC is used to classify your data,svc is a regresor here it classify whether its 0 or 1
```

```python
# training the SVM model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

Model should have above 75 % accuracy score to be accepted

```python
# accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
# accuracy score on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
```
